```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer

data = {
    'Order_ID': [1, 2, 3, 4, 5],
    'Category': ['Office Supplies', 'Furniture', 'Office Supplies', 'Technology', 'Furniture'],
    'Sales': [261.96, np.nan, 150.00, 731.94, np.nan], # قيم مفقودة تجريبية
    'Profit': [40.5, 10.2, np.nan, 150.0, 20.0]
}
df = pd.DataFrame(data)

df
```

```python
col_to_encode = 'Category'
col_index = df.columns.get_loc(col_to_encode)

dummies = pd.get_dummies(df[col_to_encode], prefix=col_to_encode)

df_dropped = df.drop(columns=[col_to_encode])

for i, new_col in enumerate(dummies.columns):
    df_dropped.insert(col_index + i, new_col, dummies[new_col])

df = df_dropped
print("--- Data after Encoding at the same position ---")
print(df.head())


```

```python
print("\n--- Missing Values Analysis ---")
missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Count': missing_count, 'Percentage': missing_percent})
print(missing_df)


imputer_mean = SimpleImputer(strategy='mean')
df['Sales'] = imputer_mean.fit_transform(df[['Sales']])

knn_imputer = KNNImputer(n_neighbors=2)
df['Profit'] = knn_imputer.fit_transform(df[['Profit']])

print("\n--- Data after Imputation ---")
print(df)
```
