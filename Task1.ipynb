```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv(r'D:\DEPI\final project\Datasets\Superstore Sales Dataset.csv')


print("--- Dataset Info ---")
print(df.info())
print("\n--- First 5 Rows ---")
print(df.head())
```

```python

print(df.isnull().sum())


print(f"\nDuplicate rows: {df.duplicated().sum()}")


print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
top_selling_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
top_selling_product.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.ylabel('Total Sales')
plt.show()
```
